```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
r=sns.load_dataset("flights")
r.head()
```

```python
r.tail()
```

```python
sns.lineplot(y="year",x="month",data=r,hue="passengers")
plt.title("LINE PLOT")
plt.show()
```

```python
r.shape
```

```python
r.isnull()
```

```python
r.isnull().sum()
```

```python
sns.heatmap(r.isnull())
plt.title("HEAT MAP")
plt.show()
```

```python
sns.barplot(data=r,x="year",y="passengers")
plt.title("BAR GRAPH")
plt.show()
```

```python
sns.scatterplot(data=r,x="year",y="passengers")
plt.title("SCATTER GRAPH")
plt.show()
```

```python
sns.catplot(data=r,x="year",y="passengers")
plt.title("CAT PLOT")
plt.show()
```
